###### Analsis of Ship Mode

```python
plt.figure(figsize = (20,5))
sns.countplot(x = "Ship Mode", data = data)
```

```python
plt.figure(figsize=(20,7))
c = sns.barplot(x= "Ship Mode", y = "Profit", data = data)
plt.title("Ship Mode Profit")
plt.xlabel("Ship Mode")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(20,7))
c = sns.barplot(x= "Ship Mode", y = "Discount", data = data)
plt.title("Ship Mode Discount")
plt.xlabel("Ship Mode")
plt.ylabel("Discount")
plt.show()
```

```python
data_shipmode = data.groupby(['Ship Mode'])[['Sales', 'Profit', 'Discount']]. mean()
data_shipmode
```

```python
figsize = (20,10)
data_shipmode.plot.pie(subplots = True, labels = data_shipmode.index, autopct = '%1.1f%%')
plt.show()
```

### Insights:

###### sales

Lowest sales: Standard Class(24.5%)

Highest sales: Same Day(25.5%)

###### Profits
